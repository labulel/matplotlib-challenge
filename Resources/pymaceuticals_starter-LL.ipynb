## Observations and Insights 

Observations:
1. There's an equal number of female vs male mice studied as shown in the pie chart;
2. Capomulin appears to be the most effective drug regimen to treat the tumor. It produced the most consistent result with narrowest IQR and smallest tumor sizes compared to other drugs (Ramicane, Infubinol, and Ceftamin)analyzed, as shown in the boxplot;
3. There is a strong positive correlation between mice weight and tumor volume, as shown in the linear regression chart.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np

# Study data files
mouse_metadata_path = "Mouse_metadata.csv"
study_results_path = "Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merge_df = pd.merge(left = mouse_metadata, right= study_results, on = "Mouse ID", how = "outer")

# Display the data table for preview
merge_df
```

```python
# Checking the number of mice.
mice_count = merge_df["Mouse ID"].value_counts()
len(mice_count)
```

```python
# Check the data for any mouse ID with duplicate time points
small_df = merge_df.loc[:,["Timepoint","Mouse ID"]]
duplicates = small_df[small_df.duplicated()]
duplicates
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

only_dup = merge_df.loc[merge_df["Mouse ID"]=="g989",:]
only_dup
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = merge_df.loc[merge_df["Mouse ID"] != "g989",:]
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
clean_count = clean_df["Mouse ID"].value_counts()
len(clean_count)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method is the most straighforward, creating multiple series and putting them all together at the end.

```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method produces everything in a single groupby function
GB_df_mean = clean_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].mean()
GB_df_median = clean_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].median()
GB_df_var = clean_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].var()
GB_df_std = clean_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].std()
GB_df_sem = clean_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].sem()
summary_df = pd.DataFrame({"Mean":GB_df_mean, "Median":GB_df_median,"Variance":GB_df_var,"Standard Deviation":GB_df_std,"SEM":GB_df_sem })
summary_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
```

```python
#Capomulin Analysis
df2 = clean_df.loc[:,["Mouse ID","Drug Regimen","Timepoint","Tumor Volume (mm3)"]]
df3_Capomulin = df2.loc[(df2["Drug Regimen"]=="Capomulin")]
df4_Capomulin = df3_Capomulin.loc[df3_Capomulin.groupby(["Drug Regimen","Mouse ID"])["Timepoint"].idxmax()]
df4_Capomulin
```

```python
quartiles_Capomulin = df4_Capomulin["Tumor Volume (mm3)"].quantile([0.25,0.5,0.75])
lowerq_Capomulin = quartiles_Capomulin [0.25]
upperq_Capomulin = quartiles_Capomulin [0.75]
iqr_Capomulin = upperq_Capomulin-lowerq_Capomulin
lower_bound_Capomulin = lowerq_Capomulin - (1.5*iqr_Capomulin)
upper_bound_Capomulin = upperq_Capomulin + (1.5*iqr_Capomulin)
print(f"LQ is {lowerq_Capomulin}, UQ is {upperq_Capomulin}, IQR of A is {iqr_Capomulin}")
print(f"Lower bond is {lower_bound_Capomulin}, Upper bond is {upper_bound_Capomulin}")
outlier_Capomulin = df4_Capomulin.loc[(df4_Capomulin["Tumor Volume (mm3)"] < lower_bound_Capomulin)|(df4_Capomulin["Tumor Volume (mm3)"] > upper_bound_Capomulin)]
outlier_Capomulin
```

```python
plt.boxplot(df4_Capomulin["Tumor Volume (mm3)"])
plt.title("Tumor Volume - Capomulin")
plt.ylabel("Tumor Volume (mm3)")
plt.ylim (20,50)
print("Observation: No outliers in Capomulin dataset")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
```

```python
correlation = st.pearsonr(df9["Weight (g)"],df9["Tumor Volume (mm3)"])
print(f"The correlation between mouse weight and average tumor volume is {round(correlation[0],2)}")
```

```python
x_values = df9["Weight (g)"]
y_values = df9["Tumor Volume (mm3)"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope, 2))+ "x + " + str(round(intercept,2))
plt.scatter(x_values, y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(21,39),fontsize=15, color="red")
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Weight and Tumor Volume Linear Regression - Capomulin")
```
